```python

xs = []
ys = []

ndata = 2200
for i in range(ndata):
    if i%10==0:
        print i
    img = Image.open(d+'/label_%d.png'%i).convert('L')
    imgr = np.asarray(Image.open(d+'/rgb_%d.png'%i).convert('RGB')).astype(np.float32).transpose(2,0,1)
    m = np.zeros((360,360,40),dtype=np.int32)
    for j in range(40):
        m[:,:,j] += np.asarray(img)==j*6
    ys.append(np.argmax(m,axis=2))
    xs.append((imgr-128.0)/128.0)
```

```python
np.savez_compressed("images.npz",np.asarray(xs))
np.savez_compressed("labels.npz",np.asarray(ys))
```
